```python
import re

def limpa_pontuacao(texto):
    # Remove ponto final ou exclamação no fim de frases, mas mantém interrogação
    texto = re.sub(r'([^.?!])\.(\s|$)', r'\1\2', texto)
    texto = re.sub(r'!\s*$', '', texto)
    return texto

```

```python
import torch
from TTS.api import TTS
import sounddevice as sd
from IPython.display import Audio
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"

print(device)

# carrega XTTS-v2 na GPU
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2",
          progress_bar=False).to(device)
```

```python
# sintetiza para NumPy (float32, faixa [-1, 1])
wav = tts.tts(
    text="No princípio criou Deus os céus e a terra, Te carai tchova lavacatai, da glória",
    speaker_wav="Voz_Nielsen.wav",
    language="pt",
)

# obter taxa de amostragem do modelo (fallback para 24000 Hz)
sr = getattr(getattr(tts, "synthesizer", None), "output_sample_rate", None) or 24000

# garantir 1D (mono) para tocar
wav = np.asarray(wav).astype(np.float32).reshape(-1)

# tocar (bloqueante até terminar)
# sd.play(wav, sr)
# sd.wait()
Audio(np.asarray(wav).reshape(-1), rate=sr, autoplay=True)
```
